$$
\mathbf{E} = \{(x_1,x_2) \in \mathbb{R}, f(x_1,x_2)=c\}=f^{-1}(\{c\})\newline

$$
\lim_{n\to +\infty}f(x_1n,x_2n)=+\infty

$$

**conclusion 2:** $E$ est borné

```python
#Fonction donnee dans l'ennonce pour afficher les courbes de niveaux sur une grille 
#Permet la verification de nos resultats
def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
    return( plt.show() )
```

```python
#fonction gradient donnee dans l'ennonce
def grad(f):
    g = autograd.grad
    def grad_f(x, y):
        return np.array([g(f, 0)(x, y), g(f, 1)(x, y)])
    return grad_f
```

```python
#fonction jacobien donnee dans l'ennonce
def J(f):
    j = autograd.jacobian
    def J_f(x, y):
        return np.array([j(f, 0)(x, y), j(f, 1)(x, y)]).T
    return J_f
```

```python
#fonction permettant d'afficher les points calculés par nos futures fonctions par dessus les courbes de niveau associées
def display_contour_pointé(f, tab, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    a, b = tab
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.scatter(a, b, color = 'r')
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
    return( plt.show() )
```

### Tâche 1 :

```python
eps = 1e-12
N=100

def Newton(F, x0, y0, eps=eps, N=N):
    P = J(F)
    for i in range(N):
        x0 = np.float(x0)
        y0 = np.float(y0)
        P_inv= np.linalg.inv(P(x0,y0))
        (x,y) = (x0,y0)-np.dot(np.linalg.inv(P(x0,y0)), F(x0,y0))
        if np.sqrt((x - x0)**2 + (y - y0)**2) <= eps:
            return x, y
        x0, y0 = x, y
       
    else:  
        raise ValueError(f"no convergence in {N} steps.")
```

### Question 6:

Afin de se déplacer vers la droite d'un point de de référence $( x_0, y_0)$, nous définissons la matrice de rotation pour un angle de $\frac{-\pi}{2}$.

En calculant le produit scalaire entre ${\nabla f(x_0,y_0)}$ et cette matrice rotation, nous obtenons un vecteur tangeant à la courbe à droite du point
$( x_0, y_0)$ (Cf dessin ci-dessous).
Afin que le point trouvé par la fonction Newton soit appliquée à une fonction de $\mathbb{R^n}\to\mathbb{R^n}$ nous définissons la fonction "fonction_adaptée"
qui renvoie cette fois $(f(x_1,x_2)-c$, $\sqrt{{x_1}^2+{x_2}^2} - \delta)$. Le second membre permet ainsi de vérifier que le point trouver
par la méthode Newton soit à une distance delta du point de référence. 


![courbe_explicative1.png](courbe_explicative1.png)

```python
def level_curve(f, x0, y0, delta, N=100, eps=eps):
    L = [[x0,y0]]
    matrice_rotation = np.array([[0,1],[-1,0]])

    def fonction_adaptée(x,y):
        return np.array([f(x,y) - f(x0, y0), np.sqrt( (x-L[-1][0])**2 + (y-L[-1][1])**2 ) - delta ])
    
    for i in range(N-1) :
        grad_F = grad(f)(x0,y0)
        
        tangente_F = matrice_rotation.dot(grad_F)
        
        tan_delta = delta*(tangente_F/np.linalg.norm(tangente_F))
        
        #on se decale du point depart car ce point correspond à un endroit ou f s'annule
        #si on demarre de ce point, Newton s'arrête au premier tour et ne cherche pas de points supplémentaires
        
        x1, y1 = np.array(x0 + tan_delta[0]), np.array(y0 + tan_delta[1])

        new_point = Newton(fonction_adaptée,x1,y1,eps,100)
        L.append(new_point)
        x0,y0= new_point
        
    #on passe par un array pour pouvoir manipuler plus facilement
    array = np.array(L)
    L = array.T

    
    #on retourne la liste au format (2,N) grace a la transpose
    return L
```

### Question 7

Nous avons implémenté une fonction de level_curve qui s'arrête lorsque le nouveau segment produit intersecte le premier
segment de la courbe de niveau.
En effet, nous avons considéré que le cas des courbes autant intersectantes était rare car il ne peut se produire que pour un
seul niveau d'une même courbe et un en seul point de cette dernière. De plus, le tracé de la courbe de niveau ne s'arrête que 
si l'on demarre autours du point de croisement.
Nous avons donc décidé de vérifier le croisement des deux derniers points tracés avec le segment formé par les deux points initiaux 
uniquement pour faciliter nos calculs.

```python
def level_curve_V2(f, x0, y0, delta, N=100, eps=eps):
    L = [[x0,y0]]
    matrice_rotation = np.array([[0,1],[-1,0]])

    def fonction_adaptée(x,y):
        return np.array([f(x,y) - f(x0,y0), np.sqrt( (x-L[-1][0])**2 + (y-L[-1][1])**2 ) - delta ])
    
    for i in range(N-1) :
        grad_F = grad(f)(x0,y0)
        
        tangente_F = matrice_rotation.dot(grad_F)
        
        tan_delta = delta*(tangente_F/np.linalg.norm(tangente_F))
        
        x1, y1 = np.array(x0 + tan_delta[0]), np.array(y0 + tan_delta[1])
        new_point = Newton(fonction_adaptée,x1,y1,eps,100)
        
        if i>=2 :
            point_departx = [L[0][0], L[1][0]]
            point_departy = [L[0][1], L[1][1]]
            ymax = max(L[-1][1], new_point[1])
            ymin = min(L[-1][1], new_point[1])
            xmax = max(L[-1][0], new_point[0])
            xmin = min(L[-1][0], new_point[0])
            for x in point_departx : 
                for y in point_departy:
                    if ymin < y < ymax and xmin < x < xmax : 
                        return(np.array(L).T)

        L.append(new_point)
        x0,y0=new_point
        
    #on passe par un array pour pouvoir manipuler plus facilement
    array = np.array(L)
    L = array.T

    
    #on retourne la liste au format (2,N) grace a la transpose
    return L
```

Grâce aux données de l'énoncé on peut déduire 8 équations vérifiées par les inconnues $a,b,c,d,e,f$ en fonction des 
coordonnées de $\vec{u_1}$ , $\vec{u_2}$, $P_1=(x_1,y_1)$ et de $P_2=(x_2,y_2)$ :

$$
a=x_1 \newline b=\alpha u_{1x} \newline c= x_2-x_1-\alpha u_{1x}\newline d=y_1 \newline e=\alpha u_{1y}\newline f=y_2-y_1 -\alpha u_{1y} \newline c=\frac{\beta u_{2x}-\alpha u_{1x}}{2} \newline f=\frac{\beta u_{2y}-\alpha u_{1y}}{2}
$$

Et après résolution de ce système on trouve:

$$
\alpha=\frac{2 [u_{2y} (x_2-x_1)+ u_{2x}(y_1-y_2)]}{u_{1x}u_{2y}-u_{1y}u_{2x}} \newline \alpha= \frac{2\det(A)}{\det(B)}
$$

où on a :

$$ A=\begin{pmatrix}  u_{2y} & (y_2-y_1)\\ u_{2x} & (x_2-x_1)\end{pmatrix} \newline $$

$$ B=\begin{pmatrix}  u_{1x} & u_{2x}\\ u_{1y} & u_{2y}\end{pmatrix} $$


On en déduit alors que les conditions pour que des solutions existent si et seulement si $\det(B)\ne0$, ce qui siginifie

```python
def gamma(t,P1,P2,u1,u2):

    det1 = u1[0]*u2[1] - u2[0]*u1[1]
    det2 = u2[1]*(P2[0]-P1[0]) - u2[0]*(P2[1]-P1[1])
    alpha = (2*det2) / det1
    
    if det1 == 0 or det2 == 0:
        #on fait un chemin lineaire quand les conditions ne sont pas respectées
        xt = (P2[0] - P1[0])*t + P1[0]
        yt = (P2[1]-P1[1])*t + P1[1]
        return np.array([xt,yt]).reshape(2,len(t))
        
    else:
        #on implémente la fonction gamma si les conditions sont respectées
        a = P1[0]
        b = alpha * u1[0]
        c = P2[0] - P1[0] - alpha*u1[0]
        d = P1[1]
        e = alpha*u1[1]
        f = P2[1]-P1[1] - alpha*u1[1]
        
        return np.array((a + b*t + c*(t**2), d +e*t + f*(t**2))).reshape(2,len(t))
```

```python
def level_curve_V3(f, x0, y0, oversampling, delta, N=100, eps=eps):
    L = [[x0,y0]]
    matrice_rotation = np.array([[0,1],[-1,0]])
    t = np.linspace(0,1,oversampling+1)

    def fonction_adaptée(x,y):
        return np.array([f1(x,y) - 0.8, np.sqrt( (x-L[-1][0])**2 + (y-L[-1][1])**2 ) - delta ])
    
    for i in range(N-1) :
        grad_F = grad(f)(x0,y0)
        
        tangente_F = matrice_rotation.dot(grad_F)
        
        tan_delta = delta*(tangente_F/np.linalg.norm(tangente_F))
        
        x1, y1 = np.array(x0 + tan_delta[0]), np.array(y0 + tan_delta[1])
        new_point = Newton(fonction_adaptée,x1,y1,eps,100)
        
        if i>=2 :
            point_departx = [L[0][0], L[1][0]]
            point_departy = [L[0][1], L[1][1]]
            ymax = max(L[-1][1], new_point[1])
            ymin = min(L[-1][1], new_point[1])
            xmax = max(L[-1][0], new_point[0])
            xmin = min(L[-1][0], new_point[0])
            for x in point_departx : 
                for y in point_departy:
                    if ymin < y < ymax and xmin < x < xmax : 
                        return(np.array(L).T)

        if oversampling > 1 :
            p1 = x0, y0
            p2 = new_point
            u1, u2 = matrice_rotation.dot(grad(f)(x0,y0)), matrice_rotation.dot(grad(f)(new_point[0],new_point[1]))
            gam = gamma(t,p1,p2,u1,u2)
            for i in range(len(gam[0])):
                L.append((gam[0][i],gam[1][i]))
        L.append(new_point)
        x0,y0=new_point
        
    #on passe par un array pour pouvoir manipuler plus facilement
    array = np.array(L)
    L = array.T
    #on retourne la liste au format (2,N) grace a la transpose
    return L
```

```python
tab10 = level_curve_V3(f1, 0.44, 0.44, 2, 0.3, N=100, eps=eps)
tab11 = level_curve_V3(f1, 0.44, 0.44, 1, 0.3, N=100, eps=eps)
```

```python
def new_display_contour3(f, tab, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    a0, b0 = tab
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    
        
    plt.scatter(a0, b0, color = 'r')
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
    return( plt.show() )
```
